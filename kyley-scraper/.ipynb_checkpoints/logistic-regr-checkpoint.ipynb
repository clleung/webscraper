```python
import pandas as pd
```

```python
lst = pd.read_html('https://en.wikipedia.org/wiki/Logistic_regression')
lst
```

```python
df = lst[2]
df
```

```python
df = df.pivot_table()
# df = df.drop(index = 0)
df
```

```python
df['Hours'] = df.Hours.astype(float)
df['Pass'] = df.Pass.astype(int)
```

```python
df.dtypes
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(penalty = 'none')
```

```python
df['Hours'] # == df.Hours
# Single bracket -> series, Double brackets -> dataframe with one column
```

```python
X = df.[['Hours']]
y = df.Pass
```

```python
clf.fit(X,y)
```

```python
clf.predict_proba([[3]])
# if studied for 3 hours, prob of passing vs prob of not passing
```
